```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_json('AOC_recent_tweets.txt')
df.set_index('id', inplace=True)

def time_in_hours(dt):
    return dt.hour + dt.minute/60 + dt.second/3600

df['created_at'] = pd.to_datetime(df['created_at'])

df['hours'] = df['created_at'].apply(time_in_hours)

df[['created_at', 'hours', 'full_text']].to_csv('AOC_tweets.csv')
```

```python
plt.figure(figsize=(12, 6))
df["hours"].plot.hist(bins=24, edgecolor="black")
plt.xlabel("Time of Day (Hours)")
plt.ylabel("Number of Tweets")
plt.title("Distribution of AOC's Tweets by Hour")
plt.xticks(range(24))
plt.grid(True)
plt.show()
```
